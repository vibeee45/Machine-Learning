```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\91958\Desktop\ML\datasets\Salary_Data.csv")

df.head()
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["YearsExperience"], df["Salary"])

plt.xlabel("Years of Experience")

plt.ylabel("Salary")

plt.title("Experience vs Salary")

plt.show()
```

```python
X = df[["YearsExperience"]]

y = df["Salary"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Data:", len(X_train))
print("Testing Data:", len(X_test))
```

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

```python
print("Slope (Coefficient):", model.coef_[0])

print("Intercept:", model.intercept_)
```

```python
y_pred = model.predict(X_test)

y_pred
```

```python
comparison = pd.DataFrame({
    "Actual Salary": y_test.values,
    "Predicted Salary": y_pred
})

comparison
```

```python
mse = mean_squared_error(y_test, y_pred)

r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)

print("R² Score:", r2)
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["YearsExperience"], df["Salary"], color="blue", label="Actual Data")

plt.plot(df["YearsExperience"],
         model.predict(df[["YearsExperience"]]),
         color="red",
         linewidth=2,
         label="Regression Line")

plt.xlabel("Years of Experience")

plt.ylabel("Salary")

plt.title("Linear Regression")

plt.legend()

plt.show()
```

```python
experience = [[5]]

predicted_salary = model.predict(experience)

print("Predicted Salary for 5 Years Experience = ₹", predicted_salary[0])
```

```python
new_experience = [[2], [4], [6], [8], [10]]

predictions = model.predict(new_experience)

result = pd.DataFrame({
    "Experience": [2, 4, 6, 8, 10],
    "Predicted Salary": predictions
})

result
```
